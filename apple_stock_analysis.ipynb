## üì¶ Installing Required Libraries

We install the `yfinance` library to fetch stock data and `matplotlib` for data visualization.


```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

## üìö Importing Libraries

We import `yfinance` for stock data, `pandas` for data manipulation, and `matplotlib` for plotting.


```python
apple = yf.Ticker("AAPL")
```

## üîç Creating a Ticker Object for Apple (AAPL)

We create a Ticker object for Apple Inc. using its stock symbol "AAPL". This allows us to access historical data and other financial information.


```python
apple_share_price = apple.history(period="5y")
apple_share_price.head()
```

## üìâ Extracting Historical Share Price Data

We use the `history()` method to fetch the last 5 years of Apple stock data, including Open, High, Low, Close, Volume, and more.


```python
plt.figure(figsize=(12, 6))
apple_share_price["Open"].plot()
plt.title("Apple Opening Stock Price - Last 5 Years")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.grid(True)
plt.show()
```

## üìä Visualizing Opening Prices

We visualize Apple's opening stock price over the last 5 years to understand its overall trend.


```python
dividends = apple.dividends
dividends.tail()
```

## üí∞ Extracting Dividend Data

We access the dividend payouts using the `dividends` attribute. These are amounts paid per share to Apple shareholders.


```python
plt.figure(figsize=(10, 5))
dividends.plot(kind='bar')
plt.title("Apple Dividend Payments Over Time")
plt.xlabel("Date")
plt.ylabel("Dividend per Share (USD)")
plt.grid(True)
plt.show()
```

## üìà Visualizing Dividends

We use a bar chart to plot Apple‚Äôs dividend payments over time, helping us see how consistently the company has rewarded shareholders.


```python
splits = apple.splits
splits.tail()
```

## ‚úÇÔ∏è Extracting Stock Splits

Stock splits affect the number of shares and their price. We extract Apple's split history using the `splits` attribute.


```python
plt.figure(figsize=(10, 5))
splits.plot(kind='bar')
plt.title("Apple Stock Splits Over Time")
plt.xlabel("Date")
plt.ylabel("Split Ratio")
plt.grid(True)
plt.show()
```

## üìä Visualizing Stock Splits

We plot any stock splits Apple has had over time to better understand structural changes in stock issuance.
